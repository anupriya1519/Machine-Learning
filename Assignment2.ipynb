## Normal Distribution

```python
mu, sigma= 0,0.1
normal= np.random.normal(mu, sigma, 1000)
```

```python
mu, std = norm.fit(normal)
print(mu,std)
plt.hist(normal, bins=25, normed=True, alpha=0.6, color='g')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)
title = "Fit results: mu = %.2f,  std = %.2f" % (mu, std)
plt.title(title)
plt.show()
```

### Here we have generated a dataset with 1000 random points which are normally distributed. After generating the data points, we fit the data to the model.
